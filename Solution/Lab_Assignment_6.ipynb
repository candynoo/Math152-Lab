```python
print("Problem1")
t = symbols('t')
s = symbols('s', positive=True)
for i in range(1,5):
    f = t**(i)
    print(integrate(f*exp(-s*t),(t,0,oo)))
print("The general form is n!/s**(i+1)")
print("The indicated result for t**10 should be 10!/s**11")
print(integrate(t**10 *exp(-s*t),(t,0,oo)),"satifies general form")
```

```python
print("Problem2")
print("part(a)")
for i in range(1,11):
    print(atan(ln(i)).evalf())
print("guessed value of limit is 1.20")

print("part(b)")    
import matplotlib.pyplot as plt
listplot = [atan(ln(i)) for i in range(1,51)]
plt.plot(listplot,'o')
print("guessed value of limit is 1.36")

print("part(c)")
n = symbols('n')
an = atan(ln(n))
L = limit(an,n, oo)
print(L)
print(L.evalf())
```

```python
print("Problem 4")
import matplotlib.pyplot as plt
from numpy import cumsum
print("part(a)")
n = symbols('n')
f = (9/10)**n /n
print(limit(f,n,oo))

print("part(b)")
nplot = [i for i in range(1,51)]
terms = [f.subs(n,i) for i in range(1,51)]
s=0
sums = cumsum(terms)
#print(sums)
plt.plot(nplot,terms,'ro',nplot,sums,'bo')

print("part(c)")
S = summation(f,(n,1,oo))
print(S)
```
