```python
#Problem 1
print("part (a)")

x = symbols('x')
y1 = sin(x)
y2 = cos(x)
x0 = nsolve(y1-y2,x,0)
print(x0)
area_x = integrate(pi*(y2**2-y1**2),(x,0,x0))
print(area_x)

print("part(b)")
y = symbols('y')
y0 = y1.subs(x,x0)
print("y0=",y0)
area_y = integrate(pi*(asin(y))**2,(y,0,y0)) + integrate(pi*(acos(y))**2,(y,y0,1))
print(area_y)
print(area_y.evalf())

print("part(c)")
area_cyl = integrate(2*pi*x*(cos(x)-sin(x)),(x,0,x0))
print(area_cyl)

                                                        
```

```python
#Problem 2
# part(a)
a,b,c,d = symbols('a,b,c,d')
f = (a*x**3 +b*x**2+c*x+d)*sqrt(1-x**2)
V = integrate(pi*f**2,(x,-1,1))
print(V)

#part(b)
import matplotlib.pyplot as plt 
g = f.subs([(a,-0.06),(b,0.04),(c,0.1),(d,0.54)])
graph = plot((g,(x,-1,1)),(-g,(x,-1,1)))
fig = graph._backend.fig
ax = graph._backend.ax
ax.set_aspect("equal")

#part(c)
vloon = V.subs([(a,-0.06),(b,0.04),(c,0.1),(d,0.54)])
print(vloon)
```

```python
#Problem 3
# part(a)
x,y,r = symbols('x,y,r')
f = sqrt(r**2-x**2)
Vhalf = integrate((r**2-x**2)*pi,(x,r/2,r))
V = 2*Vhalf
print(simplify(V))

# part(b)
print(V.subs(r,4))
```

```python
#Problem 4
# part(a)
x,y,r,R = symbols('x,y,r,R')
x_i = solve(sqrt(r**2-(x-R)**2)-y,x)
print(x_i)
r1 = x_i[0]
r2 = x_i[1]
V = 2* integrate(pi*(r2**2-r1**2),(y,0,r))
print(V)

```
